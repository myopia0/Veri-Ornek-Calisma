```python
import numpy as np
import pandas as pd
```

```python
veri=np.random.seed(42)

veri=pd.read_csv("/content/laptop_battery_health_usage.csv")
veri.head(42)
```

```python
veri.dtypes
veri.describe()
```

```python
veri[["model_year","daily_usage_hours"]].value_counts().sort_values(ascending=True)
```

```python
veri["model_year"].value_counts()
```

```python
veri[["model_year","daily_usage_hours"]].corr()
```

```python
veri.groupby("model_year")["daily_usage_hours"].mean()
```

```python
print(f'Model Yl Ortalamas: {veri["model_year"].mean()}')
print(f'Model Yl Toplam: {veri["model_year"].sum()}')
print(f'Gnlk Kullanm Ortalamas: {veri["daily_usage_hours"].mean()}')
print(f'Gnlk Kullanm Toplam:{veri["daily_usage_hours"].sum()}')
print(f'Model Yl Varyans: {veri["model_year"].var()}')
print(f'Gnlk Kullanm Varyans: {veri["daily_usage_hours"].var()}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
veri['daily_usage_hours'].value_counts().plot(kind="bar")
```

```python
x=veri["model_year"]
y=veri["daily_usage_hours"]
plt.scatter(x,y)



plt.title("Gnlk Kullanm Grafii")
plt.xlabel("Gnlk Kullanm (saat)")
plt.ylabel("Frekans")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

sns.set_style("darkgrid")
plt.show()

plt.show()

```
